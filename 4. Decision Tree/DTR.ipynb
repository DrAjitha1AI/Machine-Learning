```python
#pandas is the library which includes various functions that is assigned as pd.
import pandas as pd
```

```python
#pd as classname called read_csv which enclosed the document that is assigned to variable and called function as dataset.
dataset=pd.read_csv('50_Startups.csv')
dataset
```

```python
#change categorical state as numerical value by using get_dummies and using drop_first=True for remove the repeated/ extra dummies
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#get input and output column name by using dataset.columns
dataset.columns
```

```python
#from variable "dataset" the unpredicted datas as below is assigned as input.
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#call variable as function
independent
```

```python
#from variable "dataset" the predicting "Profit" column is assigned as output.
dependent=dataset[['Profit']]
dependent
```

```python
#from sklearn, we split trainset and testset by importing input and output datas.
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=42)
```

```python
#for DecisionTreeRegressor we assign parameter as criterion,splitter, max_features with regressor variable and then called fit to substitute the trainset-input and output datas.
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error', splitter='best', max_features=None, random_state=42)
regressor=regressor.fit(X_train,Y_train)
```

```python
#find out the tree-rootnode and other branches based on our datas in a structural form by using matplotlib.
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#assignig testset input for findout the prediction value that is called regressor as classname.
Y_pred=regressor.predict(X_test)
```

```python
#for evaluating the model by r2 score by giving the Actual value(Y_test) and predicted value(Y_pred)
from sklearn.metrics import r2_score
r_score=r2_score(Y_pred,Y_test)
r_score
```
