```python
import pandas as pd
import matplotlib as plt
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.03, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(criterion='absolute_error', n_estimators=100, max_features=None, random_state=42)
regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
for i, estimator in enumerate(regressor.estimators_[:5]):
    plt.figure(figsize=(10,6))
    tree.plot_tree(estimator,filled=True)
    plt.title(f"Tree{i+1}in RandomForesetT=Regressor")
    plt.show()
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_pred,Y_test)
r_score
```
